```python
import pandas as pd
df = pd.read_csv("games.csv")
print(len(df))
df.head()
```

```python
def result(data):
  home=data["home_club_goals"]
  away=data["away_club_goals"]
  if home > away:
    val = 1
  elif home == away:
    val = 0
  elif home < away:
    val = -1
  return val
```

```python
df['result'] = df.apply(result, axis=1)
df.result.value_counts()
```
